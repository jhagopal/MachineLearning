```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.describe()
```

Treating Outliers

```python
df[df.Cholesterol>(df.Cholesterol.mean()+3*df.Cholesterol.std())]
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df2 = df1[df.RestingBP<=(df.RestingBP.mean()+3*df.RestingBP.std())]
df2.shape
```

```python
df[df.FastingBS>(df.FastingBS.mean()+3*df.FastingBS.std())]
```

```python
df[df.MaxHR>(df.MaxHR.mean()+3*df.MaxHR.std())]
```

```python
df[df.Oldpeak>(df.Oldpeak.mean()+3*df.Oldpeak.std())]
```

```python
df3 = df2[df.Oldpeak<=(df.Oldpeak.mean()+3*df.Oldpeak.std())]
df3.shape
```

```python
df[df.HeartDisease>(df.HeartDisease.mean()+3*df.HeartDisease.std())]
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first = True)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=20)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

Using Decision Tree Classifier

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

scores = cross_val_score(DecisionTreeClassifier(), X, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
bag_model.fit(X_train, y_train)
bag_model.oob_score_
```

```python
scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

Using SVM Classifier

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(), X, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)
bag_model.fit(X_train, y_train)
bag_model.oob_score_
```

```python
scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```
